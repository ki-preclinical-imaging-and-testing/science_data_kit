```python
import pandas as pd
from pathlib import Path
from datetime import datetime

input_fname = '2025-03-17T13-31_export.csv'
df = pd.read_csv(input_fname)
```

```python
df
```

```python
df.rename(columns={
    'Path': 'filepath'
}, inplace=True)

dirs = df[df['Type'] == 'Directory'][['filepath']]
dirs = dirs[dirs['filepath'].apply(lambda x:  'histology_images' not in x.split('/')[-1])]
dirs['date'] = dirs['filepath'].apply(lambda x:  x.split('/')[-1].split(' ')[0])
dirs['name'] = dirs['filepath'].apply(lambda x:  f"HistSlides_{x.split('/')[-1].replace(' ','_')}")
dirs['label'] = 'HistologySlides'

dirs.to_csv('Laura_HistologySlides.csv')
```
